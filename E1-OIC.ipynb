## Bibliotecas.

```python
import requests
import json
import pandas as pd
import warnings
warnings.filterwarnings('ignore')
```

## Seasons

```python
seasons = {'season':[],
        'url':[]}
url = 'http://ergast.com/api/f1/seasons.json?limit=74'
response = requests.get(url)
content = json.loads(response.content)
for item in content['MRData']['SeasonTable']['Seasons']:
        seasons['season'].append(int(item['season']))
        seasons['url'].append(item['url'])
        
pd.DataFrame(seasons).to_csv('SeasonsE1.csv' , index=False)   
```

## Races

```python
races = {'season':[],
        'round':[],
        'raceid':[],
        'raceName':[],
        'circuitId':[],
        'latitud':[],
        'longitud':[],
        'country':[],
        'date':[],
        'url':[]}


for year in list(range(2010,2023)):
    url = 'http://ergast.com/api/f1/{}.json'
    response = requests.get(url.format(year))
    content = json.loads(response.content)
    for item in content['MRData']['RaceTable']['Races']:
        races['season'].append(int(item['season']))
        races['round'].append(int(item['round']))
        races['raceid'].append(item['season'] + '_' + item['round'])
        races['raceName'].append(item['raceName'])
        races['circuitId'].append(item['Circuit'].get('circuitId'))
        races['latitud'].append(float(item['Circuit'].get('Location').get('lat')))
        races['longitud'].append(float(item['Circuit'].get('Location').get('long')))
        races['country'].append(item['Circuit'].get('Location').get('country'))
        races['date'].append(pd.to_datetime(item['date']))
        races['url'].append(item['url'])

df = pd.DataFrame(races)        
pd.DataFrame(races).to_csv('RacesE1.csv' , index=False)
df
```

## Drivers

```python
drivers = {'driverId':[],
        'FirstName':[],
        'FamilyName':[],
        'DOB':[],
        'Nationality':[],
        'Biography':[]}
url = 'http://ergast.com/api/f1/drivers.json?limit=857'
response = requests.get(url)
content = json.loads(response.content)
for item in content['MRData']['DriverTable']['Drivers']:
        drivers['driverId'].append(str(item['driverId']))
        drivers['FirstName'].append(str(item['givenName']))
        drivers['FamilyName'].append(str(item['familyName']))
        drivers['DOB'].append((item['dateOfBirth']))
        drivers['Nationality'].append(str(item['nationality']))
        drivers['Biography'].append(item['url'])

df = pd.DataFrame(drivers)
pd.DataFrame(drivers).to_csv('DriversE1.csv' ,index=False)
df
```

## Constructors

```python
constructors = {'constructorId':[],
        'Name':[],
        'Nationality':[],
        'url':[]}
url = 'http://ergast.com/api/f1/constructors.json?limit=211'
response = requests.get(url)
content = json.loads(response.content)
content
for item in content['MRData']['ConstructorTable']['Constructors']:
    constructors['constructorId'].append(str(item['constructorId']))
    constructors['Name'].append(str(item['name']))
    constructors['Nationality'].append(str(item['nationality']))
    constructors['url'].append(item['url'])
    
df = pd.DataFrame(constructors)
pd.DataFrame(constructors).to_csv('ConstructorsE1.csv', index = False)
df
```

## Circuits

```python
circuits = {'circuitId':[],
        'CircuitName':[],
        'Locality':[],
        'Country':[],
        'Latitude':[],
        'Longitude':[],
        'url':[]}
url = 'http://ergast.com/api/f1/circuits.json?limit=77'
response = requests.get(url)
content = json.loads(response.content)
for item in content['MRData']['CircuitTable']['Circuits']:
        circuits['circuitId'].append(str(item['circuitId']))
        circuits['CircuitName'].append(str(item['circuitName']))
        circuits['Locality'].append(str(item['Location'].get('locality')))
        circuits['Country'].append(str(item['Location'].get('country')))
        circuits['Latitude'].append(float(item['Location'].get('lat')))
        circuits['Longitude'].append(float(item['Location'].get('long')))
        circuits['url'].append(item['url'])

df = pd.DataFrame(circuits)
pd.DataFrame(circuits).to_csv('CircuitsE1.csv', index = False)
df
```

## Qualifying

```python
qualifying = {'Year':[],
        'round':[],
        'raceId':[],
        'qualifyid':[],
        'RaceName':[],
        'driverId':[],     
        'Number':[],
        'Position':[],
        'constructorId':[],
        'Q1':[],
        'Q2':[],
        'Q3':[]}

urls = []

for i in list(range(2010,2023)):
    url1 = 'http://ergast.com/api/f1/'
    url2 = url1 + str(i)
    url3 = url2 + '.json'
    response = requests.get(url3)
    content = json.loads(response.content)
    for i in list(range(1,len(content['MRData']['RaceTable']['Races'])+1)):
        url4 = url2 + '/' +str(i) + '/qualifying.json' 
        urls.append(url4)
        
for j in urls:
    response = requests.get(j)
    content=json.loads(response.content)
    for m1 in content['MRData']['RaceTable']['Races']:
        for m2 in content['MRData']['RaceTable']['Races'][0]['QualifyingResults']:
            qualifying['Year'].append(int(m1['season']))
            qualifying['round'].append(int(m1['round']))
            qualifying['raceId'].append(m1['season'] + '_' + m1['round'])
            qualifying['qualifyid'].append('Qualifying' + '_'+ m1['season'] + '_' + m1['round'] )
            qualifying['RaceName'].append(str(m1['raceName']))
            qualifying['driverId'].append(str(m2['Driver'].get('driverId')))
            qualifying['Number'].append(int(m2['number']))
            qualifying['Position'].append(int(m2['position']))
            qualifying['constructorId'].append(str(m2['Constructor'].get('constructorId')))
            qualifying['Q1'].append(m2['Q1'])
            try:
                qualifying['Q2'].append(m2['Q2'])
            except:
                qualifying['Q2'].append('')
            try:
                qualifying['Q3'].append(m2['Q3'])
            except:
                qualifying['Q3'].append('')

df = pd.DataFrame(qualifying)
df.to_csv('Qualifying.csv', index =False)
df
```
