```python
import requests
import json
import pandas as pd
import warnings
warnings.filterwarnings('ignore')
```

## LapTimes

```python
LapTimes = {'year':[],
        'round':[],
        'RaceName':[],
        'raceId':[],
        'driverId':[],  
        'lap':[],
        'position':[],
        'time':[]}

urls = []
for i in list(range(2010,2023)):
    url1 = 'http://ergast.com/api/f1/'
    url2 = url1 + str(i)
    url3 = url2 + '.json'
    response = requests.get(url3)
    content = json.loads(response.content)
    for i in list(range(1,len(content['MRData']['RaceTable']['Races'])+1)):
        url4 = url2 + '/' +str(i) + '/laps.json?limit=1000' 
        urls.append(url4)

for j in urls:
    response = requests.get(j)
    content=json.loads(response.content)
    for m1 in content['MRData']['RaceTable']['Races']:
        for m2 in content['MRData']['RaceTable']['Races'][0]['Laps']:
            for m3 in m2['Timings']:
                LapTimes['year'].append(int(m1['season']))
                LapTimes['RaceName'].append(str(m1['raceName']))
                LapTimes['round'].append(int(m1['round']))
                LapTimes['raceId'].append(m1['season'] + '_' + m1['round'])
                LapTimes['position'].append(int(m3['position']))
                LapTimes['driverId'].append(str(m3['driverId']))
                LapTimes['lap'].append(int(m2['number']))
                LapTimes['time'].append(m3['time'])

df = pd.DataFrame(LapTimes)
df.to_csv('LapTimesE3.csv', index = False)
df
```

## DriverStandings

```python
DriverStandings = {'year':[],
        'round':[],
        'raceId':[],
        'driverStandingsId':[],
        'driverId':[],  
        'points':[],
        'position':[],
        'positionText':[],
        'wins':[]}
urls = []
for i in list(range(2010,2023)):
    url1 = 'http://ergast.com/api/f1/'
    url2 = url1 + str(i)
    url3 = url2 + '.json'
    response = requests.get(url3)
    content = json.loads(response.content)
    for i in list(range(1,len(content['MRData']['RaceTable']['Races'])+1)):
        url4 = url2 + '/' +str(i) + '/driverStandings.json?limit=1000' 
        urls.append(url4)
for j in urls:
    response = requests.get(j)
    content=json.loads(response.content)
    for m1 in content['MRData']['StandingsTable']['StandingsLists']: 
        for m2 in content['MRData']['StandingsTable']['StandingsLists'][0]['DriverStandings']:
            DriverStandings['year'].append(int(m1['season']))
            DriverStandings['round'].append(int(m1['round']))
            DriverStandings['raceId'].append(m1['season'] + '_' + m1['round'])
            DriverStandings['driverStandingsId'].append('Standings' + '_' + m1['season'] + '_' + m1['round'])
            DriverStandings['position'].append(int(m2['position']))
            DriverStandings['positionText'].append(str(m2['positionText']))
            DriverStandings['driverId'].append(str(m2['Driver']['driverId']))
            DriverStandings['points'].append(float(m2['points']))
            DriverStandings['wins'].append(int(m2['wins']))
                            
df = pd.DataFrame(DriverStandings)
df.to_csv('DriverStandingsE3.csv', index = False)
df
```

## ConstructorStandings

## ConstructorResults

```python
Data = {'position':[],
        'constructorId':[],
        'points':[],
       'year':[],
       'RaceName':[],
       'round':[],
       'raceId':[]}
DataFrame = pd.DataFrame(Data)
DataFrame = DataFrame.set_index('constructorId')
urls = []
for i in list(range(2010,2023)):
    url1 = 'http://ergast.com/api/f1/'
    url2 = url1 + str(i)
    url3 = url2 + '.json'
    response = requests.get(url3)
    content = json.loads(response.content)
    for i in list(range(1,len(content['MRData']['RaceTable']['Races'])+1)):
        url4 = url2 + '/' +str(i) + '/results.json?limit=1000' 
        urls.append(url4)
for j in urls:
    ConstructorResults = {'constructorId':[],  
                           'points':[]}
    response = requests.get(j)
    content = json.loads(response.content)
    for m1 in content['MRData']['RaceTable']['Races']:
        for m2 in content['MRData']['RaceTable']['Races'][0]['Results']:
            ConstructorResults['constructorId'].append(str(m2['Constructor']['constructorId']))
            ConstructorResults['points'].append(float(m2['points']))
        df = pd.DataFrame(ConstructorResults)
        df2 = pd.DataFrame(df.groupby('constructorId').sum().sort_values('points',ascending = False))
    RaceYear = {'RaceName':[],
                'position':[],
            'year':[],
            'round':[],
            'raceId':[]}
    for i in list(range(0,len(df2))):
        RaceYear['round'].append(m1['round'])
        RaceYear['RaceName'].append(m1['raceName'])
        RaceYear['year'].append(m1['season'])
        RaceYear['raceId'].append(m1['season'] + '_' + m1['round'])
        RaceYear['position'].append(int(i+1))

    df2['year'] = RaceYear['year']
    df2['RaceName'] = RaceYear['RaceName']
    df2['round'] = RaceYear['round']
    df2['raceId'] = RaceYear['raceId']
    df2['position'] = RaceYear['position']
    DataFrame =  pd.concat([DataFrame,df2])

DataFrame.to_csv('ConstructorResultsE3.csv', index = True)
DataFrame
```

### Otra forma para construir esta tabla es directamente  desde la tabla results

```python
Results = {'year':[],
        'raceId':[],
        'RaceName':[],
        'resultId':[],
        'driverId':[],
        'constructorId':[],
        'number':[],
        'grid':[],     
        'position':[],
        'positionText':[],
        'points':[],
        'laps':[],
        'time':[],
        'miliseconds':[],
        'fastestLap':[],
        'rank':[],
        'fastestLapTime':[],
        'fastestLapSpeed(kph)':[],
        'Status':[]}
urls = []
for i in list(range(2010,2023)):
    url1 = 'http://ergast.com/api/f1/'
    url2 = url1 + str(i)
    url3 = url2 + '.json'
    response = requests.get(url3)
    content = json.loads(response.content)
    for i in list(range(1,len(content['MRData']['RaceTable']['Races'])+1)):
        url4 = url2 + '/' +str(i) + '/results.json?limit=1000' 
        urls.append(url4)
for j in urls:
    response = requests.get(j)
    content=json.loads(response.content)
    for m1 in content['MRData']['RaceTable']['Races']:
        for m2 in content['MRData']['RaceTable']['Races'][0]['Results']:
            Results['year'].append(int(m1['season']))
            Results['raceId'].append(m1['season'] + '_' + m1['round'])
            Results['RaceName'].append(str(m1['raceName']))
            Results['resultId'].append('Result' + '_' + m1['season'] + '_' + m1 ['round'] + '_' + m2['position'])
            Results['position'].append(int(m2['position']))
            Results['positionText'].append(str(m2['positionText']))
            Results['driverId'].append(str(m2['Driver']['driverId']))
            Results['number'].append(int(m2['number']))
            Results['constructorId'].append(str(m2['Constructor']['constructorId']))
            Results['grid'].append(int(m2['grid']))
            Results['points'].append(float(m2['points']))
            Results['laps'].append(int(m2['laps']))
            try:
                Results['time'].append(m2['Time']['time'])
            except:
                Results['time'].append('')
            try:
                Results['miliseconds'].append(m2['Time']['millis'])
            except:
                Results['miliseconds'].append('')
            try:
                Results['fastestLap'].append(m2['FastestLap']['lap'])
            except:
                Results['fastestLap'].append('')
            try:
                Results['rank'].append(m2['FastestLap']['rank'])
            except:
                Results['rank'].append('')
            try:
                Results['fastestLapTime'].append(m2['FastestLap']['Time']['time'])
            except:
                Results['fastestLapTime'].append('')
            try:
                Results['fastestLapSpeed(kph)'].append(m2['FastestLap']['AverageSpeed']['speed'])
            except:
                Results['fastestLapSpeed(kph)'].append('')
            Results['Status'].append(str(m2['status']))
            
        
df = pd.DataFrame(Results)
ConstructorResults=df.groupby(["year","RaceName",'constructorId','Status']).agg({'points':"sum"})
ConstructorResults
        
        
        
```
